# Attrition Data Analysis

### Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Reading

```python
df=pd.read_csv('Attrition data.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['NumCompaniesWorked', 'TotalWorkingYears', 'EnvironmentSatisfaction','JobSatisfaction','WorkLifeBalance'])
df.isnull().sum()
```

```python
df2=df.head(400)
df2=df2.copy()
df2
```

```python
df.describe()
```

### EDA

```python
attrition_distribution = df['Attrition'].value_counts()

labels = attrition_distribution.index  
sizes = attrition_distribution.values 
colors = ['#ff9999', '#66b3ff']  
plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title("Attrition Distribution")
plt.show()
```

```python
df.loc[:, 'TenureRatio'] = df['YearsAtCompany'] / df['TotalWorkingYears']
df[['YearsAtCompany', 'TotalWorkingYears', 'TenureRatio']].head()
```

```python
categorical_features = ['Gender', 'JobRole', 'Department', 'MaritalStatus','YearsAtCompany']
for features in categorical_features:
    plt.figure(figsize=(15, 10))
    ax = sns.countplot(x=features, data=df, hue='Attrition')
    plt.title(f'{features} vs Attrition')
    plt.xlabel(features)        
    plt.ylabel('Count')        
    
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='baseline', 
                    fontsize=12, color='black', xytext=(0, 5),
                    textcoords='offset points')
    
    plt.show()
```

```python
sns.histplot(df['Age'],kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
attrition_performance = df.groupby(['Attrition', 'PerformanceRating']).size().unstack(fill_value=0)

# Plotting
plt.figure(figsize=(10, 6))
attrition_performance.plot(kind='bar', stacked=True, colormap='coolwarm', ax=plt.gca())
plt.title('Attrition vs Performance Rating')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.legend(title='Performance Rating')
plt.show()
```

```python
attrition_counts = df[df['Attrition'] == 'Yes'].groupby('PerformanceRating').size()
total_counts = df.groupby('PerformanceRating').size()
attrition_rate = (attrition_counts / total_counts).fillna(0) * 100

plt.figure(figsize=(5, 5))
sns.barplot(x=attrition_rate.index, y=attrition_rate.values, palette='coolwarm',hue=attrition_rate.index)
plt.title('Attrition Rate by Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Attrition Rate (%)')
plt.ylim(0, max(attrition_rate.values) + 5)

for i, rate in enumerate(attrition_rate.values):
    plt.text(i, rate + 0.5, f'{rate:.1f}%', ha='center')

plt.show()
```

```python
sns.boxplot(x='Attrition',y='MonthlyIncome',data=df2)
plt.title('Attrition vs Monthly Income')
plt.show()
```

```python
sns.boxplot(x='JobRole',y='MonthlyIncome',hue='Attrition',data=df)
plt.xticks(rotation=90)
plt.title('Job role vs Monthly Income vs Attrition')
plt.show()
```

```python
sns.boxplot(x='Attrition',y='Age',hue='Attrition',data=df)
plt.xticks(rotation=90)
plt.title('Age vs Attrition')
plt.show()
```

```python
corr_matrix=df[['Age','MonthlyIncome','YearsAtCompany','TotalWorkingYears']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('correlation matrix')
plt.show()
```
